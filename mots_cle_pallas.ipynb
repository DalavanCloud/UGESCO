```python
geo = pd.read_csv(r"C:\Users\Boulot\Desktop\BelgicaPress\communes_et_sections_geonames.csv", encoding="utf8")
lieux_be = set([str(name).strip().lower() for name in geo.alternate_name])
lieux_be
```
