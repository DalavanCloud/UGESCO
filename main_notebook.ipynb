```python
file = "data/jsondata_ugesco.json"
data = json_to_df(file)
data.columns
```

```python
data.head(20)
```

```python
#Drop temporal columns
to_drop = [c for c in data.columns if c.startswith('temporal')]
data.drop(to_drop, axis=1, inplace=True)
data.head()
```

```python
#remove class and prob 3 to 5
data.drop(list(data.filter(regex = '(_class|_prob)[3-5]$')), axis = 1, inplace = True)
```

```python
#Delete "indoor", "outdoor" etc in class names and replace undescores with spaces
data['imageclassification_class1'] = data['imageclassification_class1'].str.replace("/.+", "").str.replace('_', ' ')
data['imageclassification_class2'] = data['imageclassification_class2'].str.replace("/.+", "").str.replace('_', ' ')
data.head()
```

```python
class_places = pd.read_csv("data/classes_places.csv", encoding="utf8")
class_places.head()
```

```python
# join/merge with classes_places.csv to get the class_ids
data = pd.merge(data, class_places, how='left', left_on=['imageclassification_class1'], right_on=['original_class'])
data = data.rename(columns={'class_id': 'class_id1'})
```

```python
data = pd.merge(data, class_places, how='left', left_on=['imageclassification_class2'], right_on=['original_class'], suffixes=['', '_x'])
data = data.rename(columns={'class_id': 'class_id2'})
```

```python
to_drop = [c for c in data.columns if c.endswith('_x') or c.endswith('_y') or c.startswith('AAT')]
data.drop(to_drop, axis=1, inplace=True)
data.head()
```

```python
#join/merge with pallas_photothèque to get the keywords
phototeque = pd.read_csv("data/phototheque_pallas.csv", encoding="utf8", dtype={'beeldid': str})
#phototeque.head()
data = pd.merge(data, phototeque, how='left', on=['beeldid'], suffixes=['', '_x'])
data.head()
```

```python
#join/merge with pallas_motcles_lieux_complet.csv to get the locations in the keywords
mots_cles = pd.read_csv("data/pallas_motcles_lieux_complet.csv", encoding="utf8")
#mots_cles.head()
data = pd.merge(data, mots_cles, how='left', left_on=['keywords_fr'], right_on=['keyword'], suffixes=['', '_x'])
data.head()
```

```python
#filter rows with non null value in loc_name
locs = data[data['loc_name'].notnull()]
locs.shape
```
