```python
column_names = ['beeldid','spatial_value', 'spatial_key', 'class1', 'prob1', 'class2', 'prob2']
data = pd.read_csv("data/jsondata.csv", encoding="utf8", names = column_names, dtype={'beeldid': str})
```

```python
data.head(20)
```

```python
#remove "indoor" "outdoor" etc in class1 and class2 and replace undescores with spaces
data['class1'] = data['class1'].str.replace("/.+", "").str.replace('_', ' ')
data['class2'] = data['class2'].str.replace("/.+", "").str.replace('_', ' ')
data.head()
```

```python
# join/merge with classes_places.csv to get the class_ids
class_places = pd.read_csv("data/classes_places.csv", encoding="utf8")
class_places.head()
```

```python
data = pd.merge(data, class_places, how='left', left_on=['class1'], right_on=['original_class'])
data = data.rename(columns={'class_id': 'class_id1'})
data.head()
```

```python
data = pd.merge(data, class_places, how='left', left_on=['class2'], right_on=['original_class'], suffixes=['', '_x'])
data = data.rename(columns={'class_id': 'class_id2'})
```

```python
to_drop = [c for c in data.columns if c.endswith('_x') or c.startswith('AAT')]
data.drop(to_drop, axis=1, inplace=True)
data.head()
```
